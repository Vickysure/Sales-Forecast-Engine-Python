# SALES FORECASTING ENGINE FOR BRIGHTMART RETAIL LIMITED

```python
# The details below are the data provided for the weekly sales Report of Brightmart Retail Ltd: Task 1
#this analysis is for product 1 : wireless Mouse
product_name = "Wireless Mouse"
selling_price_per_unit = selling_price = 12500
units_sold_this_week = units_sold = 180
weekly_sales_target = 200
forecast_growth_rate = 15
current_stock = 250
minimum_stock_level = 120
product_available = True

# the code below is to determine the revenue generated this week: Task 2
revenue = selling_price * units_sold
print ("Revenue Generated This Week:", revenue)

# the code below is to determine the expected sales next week based on the forecast growth rate: Task 3
next_week_forecast_sales = forecast_sales = units_sold + (units_sold * (15/100))
print ("Forecast Sales:", forecast_sales)

# the code below is to display the expected revenue for next week based on the forecasted sales: Task 4
forecast_revenue = forecast_sales * selling_price
print ("Forecast Revenue:", forecast_revenue)

# the code below is to display the update on the current stock: Task 5
current_stock -= units_sold
updated_stock = current_stock
print ("Updated Stock Quantity For the New Week:", updated_stock)

# the code below is to determine if the sales target was achieved: Task 6
target_sales_achieved = (units_sold >= weekly_sales_target)
print ("Target Sales Achieved:", target_sales_achieved)

# the code below is to determine if restocking is required: Task 7
restocking_required = (updated_stock < minimum_stock_level)
print ("Restocking Required:", restocking_required)

# the code below is to determine whether the product is ready for promotion: Task 8
product_qualifies_for_promotion = product_available and target_sales_achieved
print ("Product is Ready for Promotion:", product_qualifies_for_promotion)

# the code bewlo is to determine if the product demand is High : Task 9
product_in_high_demand = forecast_sales > 200 or forecast_revenue > 2500000
print ("Product is in High Demand:", product_in_high_demand)

#the code below is to generate the weekly sales report: Task 10
print ("BRIGHTMART SALES FORECAST REPORT 1")
print ("Product Name:", product_name)
print ("Product Selling Price Per Unit:", selling_price)
print ("Units Sold this Week:", units_sold)
print ("Weekly Revenue:", revenue)
print ("Forecast Growth Rate:", forecast_growth_rate)
print ("Forecast Sales:", forecast_sales)
print ("Forecast Revenue:", forecast_revenue)
print ("Current Stock:", updated_stock)
print ("Sales Target Achieved:", target_sales_achieved)
print ("Restocking Required:", restocking_required)
print ("Product Available:", product_available)
print ("Promotion Eligible:", product_qualifies_for_promotion)
print ("High Demand Forcast:", product_in_high_demand)
```
